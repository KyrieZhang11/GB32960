## 载入工具包

```python
import pandas as pd
# 载入自定义工具包
import package.gb32960 as gb
```

## 读取数据

```python
# 文件路径
filepath = r'source\LDPGABAC7JN200663_2019-12-30.csv'

# 打开文件
df_o = pd.read_csv(filepath, encoding='ANSI')
df_o.head()
```

## 数据治理

```python
# 初始化

# 登录
df_denglu01 = pd.DataFrame()
# 登录
df_denglu04 = pd.DataFrame()
# 登录
df_denglu05 = pd.DataFrame()
# 登录
df_denglu06 = pd.DataFrame()
# 工况
df_gongkuang = pd.DataFrame()

# 错误报文
df_error = pd.DataFrame()
list_error = []
```

```python
for i in range(len(df_o)):
    print("\n=========================",i)
    print(df_o['报文内容'][i])
    model = ['01','02','05','06','07','08','09']
    try:
        data = gb.gb32960(df_o['报文内容'][i], model)
        if data.m == '01':
            df_denglu01.append(data.pl, ignore_index=True)
        elif data.m in ['02','03']:
            df_gongkuang.append(data.pl, ignore_index=True)
        elif data.m == '04':
            df_denglu04.append(data.pl, ignore_index=True)
        elif data.m == '05':
            df_denglu05.append(data.pl, ignore_index=True)
        elif data.m == '06':
            df_denglu06.append(data.pl, ignore_index=True)
    except:
        print('第',i+1,'行有误，文本为:',df_o['报文内容'][i])
        list_error.append(i)
        error_i = pd.DataFrame({'行':[i+1], '报文内容':[df_o['报文内容'][i]]})
        df_error.append(error_i, ignore_index=True)
        
print('===== all done =====')
```
